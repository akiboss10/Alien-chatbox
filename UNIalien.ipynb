```python
import random
import re

class ChatbotApp:
    def __init__(self):
        self.chatbot_responses = {
            'greet': r'\b(hi|hello|hey|good\s*morning|good\s*evening)\b',
            'who_are_you': r'.*\b(who\s*are\s*you|tell\s*me\s*about\s*yourself)\b.*',
            'universe': r'.*\b(universe|space|galaxy|cosmos)\b.*',
            'stars': r'.*\b(stars|constellations|star\s*system)\b.*',
            'moon': r'.*\b(moon|lunar|satellite)\b.*',
            'solar_system': r'.*\b(solar\s*system|planets|sun)\b.*',
            'alien': r'.*\b(alien|extraterrestrial|ET|UFO)\b.*',
            'black_hole': r'.*\b(black\s*hole|event\s*horizon)\b.*',
            'parallel_universe': r'.*\b(parallel\s*universe|multiverse)\b.*',
            'no': r'.*\b(no|don\'t\s*know|i\s*don\'t\s*know|nope|not\s*really)\b.*',  # Handle 'no' responses
        }

        self.exit_commands = ("quit", "exit", "bye", "goodbye")
        self.user_name = ""
        self.first_interaction = True  # Flag to track the first interaction
        self.greet_bot()

    def greet_bot(self):
        # Ask for the user's name and give a compliment
        self.user_name = input("Alienbot: What's your name? ")
        compliment = random.choice([
            f"{self.user_name}, what a beautiful name you have!",
            f"Wow, {self.user_name}, your name sounds so unique!",
            f"{self.user_name}, that's a name I won't forget!",
            f"I love your name, {self.user_name}, it has a nice ring to it!"
        ])
        print(f"Alienbot: {compliment}")

        # Provide instructions on how to quit the conversation
        print("Alienbot: You can ask me about the universe, black holes, nebulae, parallel universes, time travel, dark matter, and more!")
        print("Alienbot: If you want to quit the chat, just type 'quit', 'exit', 'bye', or 'goodbye'.")

    def handle_user_input(self, user_input):
        user_input = user_input.strip()
        if user_input:
            print(f"You: {user_input}")
            if user_input.lower() in self.exit_commands:
                print("Alienbot: Goodbye! Talk to you soon.")
                return False
            elif re.search(self.chatbot_responses['no'], user_input.lower()):
                # If the user says "no" or something similar, reply with a defined message
                print("Alienbot: Oh, I see! You don't know that thing. Don't worry, we're all still learning about the vast universe!")
                return True
            else:
                bot_response = self.generate_response(user_input.lower())
                print(f"Alienbot: {bot_response}")
        return True

    def generate_response(self, user_input):
        for intent, pattern in self.chatbot_responses.items():
            if re.search(pattern, user_input):
                if intent == 'greet':
                    return random.choice(["Hello there!", "Hi! How's it going?", "Hey!"])
                elif intent == 'who_are_you':
                    return random.choice([
                        "I am Alienbot, an AI from a distant world, created to chat and learn from humans.",
                        "I'm just a friendly AI who loves to learn, share knowledge, and help humans!",
                        "I'm your AI companion from a galaxy far, far away. I'm here to assist you in any way I can."
                    ])
                elif intent == 'universe':
                    return random.choice([
                        "The universe is a vast and infinite expanse of stars, galaxies, and cosmic wonders. It's hard to comprehend how much we don't know about it. We exist as mere specks in the grand cosmic ocean. And who knows? The universe might be teeming with life, waiting to be discovered.",
                        "Imagine the universe as a giant puzzle that we've only begun to piece together. Its vastness makes us feel both insignificant and deeply connected to everything around us. Every star, planet, and galaxy holds its own mysteries, and perhaps one day, we will uncover them all.",
                        "The universe is a place of both awe and mystery, filled with celestial wonders that can captivate our imagination. From the birth of stars to the death of galaxies, the cosmos never stops changing. Its vastness reminds us how little we truly know about our place in it."
                    ])
                elif intent == 'stars':
                    return random.choice([
                        "Stars are the ultimate symbols of the universe’s mystery and beauty. They are born in the darkest corners of space, where dust and gas collide in violent fury, eventually igniting into brilliant beacons of light. Their radiance can travel for billions of years, reaching us as faint twinkles in the sky, each carrying a story of its own.",
                        "Stars are fascinating! They are born from nebulae, clouds of gas and dust, that collapse under their own gravity, eventually igniting nuclear fusion. These celestial bodies provide the energy that fuels life, and each one is part of a larger cosmic dance in the vastness of space. The closest star to Earth is, of course, the Sun, the lifeblood of our planet.",
                        "Stars are like cosmic puzzles, each one with its own life cycle. They begin their lives as clouds of gas and dust, then, through gravity and pressure, they ignite into blazing balls of nuclear fusion. The end of their life, however, is just as fascinating. Some stars explode in supernovae, scattering elements that become the building blocks of new stars and planets."
                    ])

                elif intent == 'moon':
                    return random.choice([
                        "The Moon, Earth's natural satellite, has been a constant companion to us since time immemorial. Its phases guide the tides and inspire poets, dreamers, and scientists alike. There’s a deep mystery about the Moon, especially when you think about its impact on our oceans, its role in the formation of Earth, and the fact that we’ve set foot on its surface.",
                        "The Moon has always had a magnetic pull on human imagination. It influences the Earth’s tides, stabilizes our planet’s axial tilt, and is the reason for eclipses. It’s thought to have formed when a giant object collided with Earth, sending debris into orbit, which eventually coalesced into the Moon. Today, we continue to explore its surface, with dreams of one day establishing permanent colonies there.",
                        "The Moon has inspired countless legends, myths, and scientific investigations. As Earth’s only natural satellite, it has played an essential role in the evolution of our planet. Some believe that the Moon might have even helped stabilize Earth’s axial tilt, making it more hospitable to life. It's our companion in the night sky, guiding sailors and lovers alike with its cool, silver glow."
                    ])
                elif intent == 'solar_system':
                    return random.choice([
                        "The solar system is a fascinating region of space, teeming with diversity. At the center, the Sun reigns, providing the warmth and energy that fuels life on Earth. Around it, eight planets orbit in a cosmic ballet, each with its own unique features. From the fiery surface of Mercury to the icy rings of Saturn, our solar system is a miniature universe of its own.",
                        "Our solar system is more than just a group of planets; it’s a dynamic and ever-changing environment. The Sun, a massive ball of burning gas, drives the orbits of the planets, comets, and asteroids that circle around it. The planets themselves are worlds of wonder, each one with its own mysteries, from the red deserts of Mars to the frigid clouds of Neptune.",
                        "The solar system is our cosmic neighborhood, consisting of the Sun and everything bound to it by gravity. There are eight planets, moons, asteroids, and comets, all moving through space in predictable patterns. Each planet is unique in its composition and atmosphere, and together, they form a system that’s a tiny part of the vast and mysterious universe."
                    ])
                elif intent == 'alien':
                    return random.choice([
                        "The question of extraterrestrial life has fascinated humanity for centuries. What if aliens have been observing us for millennia? Perhaps they’re waiting for the right moment to make contact. Or maybe they’re already here, disguised among us, studying us in ways we can’t even comprehend.",
                        "Could we be alone in the universe, or are there other civilizations out there? The search for alien life is one of the most exciting areas of scientific exploration. Some scientists believe that extraterrestrial life might be found in the most unexpected places, from the depths of icy moons to the distant reaches of the galaxy.",
                        "The possibility of extraterrestrial life is one of the most intriguing mysteries of the cosmos. With billions of stars in our galaxy alone, each potentially hosting planets that could support life, it seems unlikely that we are the only intelligent beings in the universe. One day, we might just find a neighbor beyond the stars!"

                    ])

                elif intent == 'black_hole':
                    return random.choice([
                        "Black holes are one of the most mysterious phenomena in the universe. They are regions of space where gravity is so strong that not even light can escape. Imagine a star collapsing under its own weight, creating an intense gravitational pull that devours everything nearby. Once you pass the event horizon, there’s no turning back. Black holes warp the very fabric of space and time, bending them in ways that challenge our understanding of the universe. Some even theorize that they might lead to other universes or dimensions.",
                        "Black holes are often described as the universe’s ultimate mystery. Once a star reaches the end of its life cycle, it can collapse under its own gravity, forming a black hole. Inside a black hole, gravity is so strong that it distorts space and time itself. The event horizon, the point of no return, marks the boundary where not even light can escape. What happens inside a black hole? That’s still one of the biggest questions in astrophysics. Some scientists believe black holes might be gateways to other dimensions or even time itself.",
                        "Black holes are fascinating yet terrifying objects in space. They are created when massive stars run out of fuel and collapse under their own gravity. Once they form, they suck in everything nearby, even light, and warp the fabric of space-time. Some scientists believe black holes may hold the key to understanding the nature of the universe. Others speculate that they might be portals to other worlds or even other times, making them one of the most intriguing and perplexing phenomena in the cosmos."
                    ])

                elif intent == 'parallel_universe':
                    return random.choice([
                       "Parallel universes, also known as the multiverse, are a concept that suggests that there are many different realities coexisting alongside our own. These alternate universes could be similar to ours or vastly different, with entirely new laws of physics or alternate versions of history. Some scientists speculate that if we could travel between these parallel universes, we might find versions of ourselves living entirely different lives. It's a fascinating and complex idea that challenges our understanding of reality and could revolutionize our knowledge of existence itself.",
                       "The idea of parallel universes is one of the most mind-boggling concepts in modern physics. Imagine an infinite number of realities, each one slightly different or even completely unique. Some theories suggest that every decision we make creates a new branch in the multiverse, resulting in countless versions of ourselves living out different lives. Could there be a universe where you made a different choice? Perhaps one where dinosaurs never went extinct or where humans never developed technology? Exploring the possibility of parallel universes could offer profound insights into the nature of existence.",
                        "Parallel universes are a mind-bending concept that suggests there could be infinite versions of reality. Some scientists propose that our universe is just one of many, each with its own unique laws of physics, history, and outcomes. This idea, often referred to as the multiverse, offers fascinating possibilities—what if there’s a universe where time flows backward or where gravity doesn’t exist? Could alternate versions of ourselves be living in these universes, making different choices and experiencing different outcomes? The idea of parallel universes challenges our conventional understanding of reality and raises deep questions about the very fabric of existence."
])


                elif intent == 'no':
                    return random.choice([
                        "Oh, I see! You don't know that thing. Don't worry, we're all still learning about the vast universe!",
                        "No? That's okay, my knowledge of the universe is vast, but I don't know everything either!",
                        "I understand! It's a big universe, and sometimes it takes time to learn all its wonders."
                    ])
        return random.choice([
            "That's interesting! Can you tell me more?",
            "I see! What else is on your mind? Or would you like to ask about black holes, time travel, or parallel universes?"
        ])

if __name__ == "__main__":
    app = ChatbotApp()
    while True:
        user_input = input("You: ")
        if not app.handle_user_input(user_input):
            break
```
